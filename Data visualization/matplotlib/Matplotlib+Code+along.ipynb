##### Bar Chart - Plot sales across each product category

- A bar chart uses bars to show comparisons between categories of data.
- A bar graph will always have two axis. 
- One axis will generally have numerical values or measures, 
- The other will describe the types of categories being compared or dimensions.

```python
import numpy as np
import matplotlib.pyplot as plt

product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52] )

plt.bar(product_category, sales)
plt.show()
```

##### Bar Chart - Plot sales across each product category

1. Adding labels to Axes
2. Reducing the bar width
3. Giving Title to the chart
4. Modifying the ticks to show information in (million dollars)

```python
import numpy as np
import matplotlib.pyplot as plt

product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52] )

# plotting bar chart and setting bar width to 0.5 and aligning it to center
plt.bar(product_category, sales, width= 0.5, align='center', edgecolor='Orange',color='cyan')

# Adding and formatting title
plt.title("Sales Across Product Categories\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Product Category", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

# Modifying the ticks to show information in (million dollars)
ticks = np.arange(0, 6000000, 1000000)
labels = ["{}M".format(i//1000000) for i in ticks]
plt.yticks(ticks, labels)

plt.show()
```

##### SubPlots - Plot Sales for various markets for years 2012 to 2015

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array(['2012', '2013', '2014', '2015'])

sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

##### Subplots are shown in the same graph as line charts, identified by different colours

```python
fig, ax = plt.subplots()  # fig is figure and ax is axis.Its like a blank canvus or blank plot.

europe, = ax.plot(years, sales_Europe) #ax.plot will return 2 values and we are only interested on one value hence mentioning europe, year will be plotted on x axis and sales_europe will be plotted on y axis.
europe.set_label("Europe")

usca, = ax.plot(years, sales_USCA)
usca.set_label("USCA")
usca.set_dashes([2, 2, 2, 2]) # 2 point lines  2 point break again 2 point lines 2 point break 

africa, = ax.plot(years, sales_africa)
africa.set_label("Africa")
africa.set_dashes([2, 2, 5, 2]) # 2 point lines  2 point break again 5 point lines 2 point break 

asia, = ax.plot(years, sales_Asia_Pacific)
asia.set_label("Asia Pacific")
asia.set_dashes([2, 2, 10, 2]) # 2 point lines  2 point break again 10 point lines 2 point break 

latam, = ax.plot(years, sales_LATAM)
latam.set_label("LATAM")
latam.set_dashes([2, 5, 5, 2]) # 2 point lines  5 point break again 5 point lines 2 point break 

plt.legend()

plt.legend(bbox_to_anchor=(1.31,0.4)) # reposition the legend anchor position

plt.show() #By default plot will change color based on the number of plot .
```

##### Subplots are shown in seperate graphs. Each chart can be of different types.

```python
fig, ax = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True)

europe, = ax[0][0].plot(years, sales_Europe) #setting 1st quadrent of plot using line plot
europe.set_label("Europe")

usca = ax[0][1].bar(years, sales_USCA) #setting 2nd quadrent of plot using bar plot
usca.set_label("USCA")

africa = ax[0][2].scatter(years, sales_africa) #setting 3rd quadrent of plot using scatter plot
africa.set_label("Africa")

asia, = ax[1][0].plot(years, sales_Asia_Pacific) #setting 4th quadrent of plot using line plot but with diffrent dashes.
asia.set_label("Asia Pacific")
asia.set_dashes([2, 2, 10, 2])

latam, = ax[1][1].plot(years, sales_LATAM)  #setting 5th quadrent of plot using line plot but with diffrent size of dashes.
latam.set_label("LATAM")
latam.set_dashes([2, 5, 5, 2])


plt.show()
```

```python
fig, ax = plt.subplots(ncols=4, sharey=True) # dividing into 4 column

europe, = ax[0].plot(years, sales_Europe) #1st column plot for europe region sales which is a line chart.
europe.set_label("Europe")
europe.set_color("red")
ax[0].set_title('Sales in Europe') 

usca = ax[1].bar(years, sales_USCA) #second column plot is bar chart and setting lebels USCA
usca.set_label("USCA")
ax[1].set_title('Sales in USCA')

africa = ax[2].scatter(years, sales_africa) #Third column is Africa sales as scatter plot 
africa.set_label("Africa")
ax[2].set_title('Sales in Africa')

# In this section we are going to combine 2 diffrent region sales plot to a single column
asia = ax[3].bar(years, sales_Asia_Pacific, width = 0.5, color='royalblue') #plotting asia region into bar chart as royal blue in 4th column

latam = ax[3].bar(years, sales_LATAM, width = 0.5, color='seagreen') #Plotting LATAM region in 4th column as seagreen.
#this will club both bar chart in same column and merge.

ax[3].set_title('Sales in Asia Pacific and LATAM')  #Setting Title Of the plot As Asia Pacific and LATAM
ax[3].legend( (asia[0], latam[0]), ('Asia Pacific', 'LATAM') ) # Setting Legand to identify which region is ploted in which color

fig.set_size_inches(20.5, 5.5, forward=True) #setting front size in inches to make plot (fig) size bigger

plt.show()
```
