```python
Let’s say you work as an analyst at UpGrad and you want to find interesting patterns in ‘the typical dates and times when UpGrad students submit their assignments’.

The data set provided below contains information on the dates and times of students’ 
submissions of an assignment. It has two variables — “submission” (the unique URL of a student’s submission on the platform) and “submit_time” (the date-time of submission).
```

```python
grades=pd.read_csv("grades.csv")
```

```python
grades
```

```python
grades["Format"]=grades["submission"].apply(lambda x: 1 if '.zip' in x else 0)
```

```python
(grades['Format'].sum()/grades.submission.count())*100
```

```python
""" How many students submitted the assignment after the first deadline (including the students who submitted after the second deadline) ? """
```

```python
grades['submit_time']=pd.to_datetime(grades['submit_time'],format='%m/%d/%y-%H:%M:%S')
```

```python
grades['day'] = grades['submit_time'].dt.day
grades['month'] = grades['submit_time'].dt.month
grades['year'] = grades['submit_time'].dt.year
grades['hour'] = grades['submit_time'].dt.hour
grades['minute'] = grades['submit_time'].dt.minute
grades['second'] = grades['submit_time'].dt.second
grades['date'] = grades['submit_time'].dt.date
```

```python
grades
```

```python
comparison_date = pd.to_datetime("01/03/17-20:18:21", format='%m/%d/%y-%H:%M:%S')
grades["is_before"]=grades["submit_time"]>comparison_date
```

```python
grades["is_before"].value_counts()
```

```python
comparison_date = pd.to_datetime("01/03/17-20:18:21", format='%m/%d/%y-%H:%M:%S')
grades["is_before"]=grades["submit_time"]>comparison_date
```
