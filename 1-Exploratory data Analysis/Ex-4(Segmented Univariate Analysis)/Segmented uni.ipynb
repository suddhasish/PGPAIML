```python
allgrp=censusdf[(censusdf["Area Name"]=="INDIA")&(censusdf["Total/Rural/Urban/"]=="Total")]
```

```python
allgrp.head(20)
```

```python
allgrp["literacy_rate"] = allgrp["Literate Persons"]/allgrp["Total Persons"]*100
```

```python
allgrp["Age-group"]=allgrp["Age-group"].astype("string")
```

```python
plt.figure(figsize=[30,10])
plt.bar(allgrp["Age-group"],allgrp["literacy_rate"])
plt.show()
```
