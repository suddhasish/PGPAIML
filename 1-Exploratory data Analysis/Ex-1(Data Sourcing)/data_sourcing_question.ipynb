```python
cendf2=cendf1[~(cendf1["Age-group"].isin(["Age not stated","All ages","0-6"]))]
```

```python
cendf2
```

```python
cendf3=cendf2[((cendf2["Area Name"]=="INDIA")&(cendf2["Total/Rural/Urban/"]=="Total"))]
```

```python
cendf3
```

```python
cendf4=cendf3.groupby(["Area Name"]).agg({'Total Persons':np.sum,'Literate Persons':np.sum})
```

```python
cendf4['Literacy Percentage'] = cendf4['Literate Persons']/cendf4['Total Persons']*100
```
